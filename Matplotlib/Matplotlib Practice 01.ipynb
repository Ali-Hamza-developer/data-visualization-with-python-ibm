```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)           # Plot x vs y
plt.show()               # Display the plot
```

```python
# 📌 Adding Title and Labels
plt.plot(x, y)           
plt.title("Simple Line Plot")     # Add a title to the graph
plt.xlabel("X Axis Label")         # Label for X-axis
plt.ylabel("Y Axis Label")         # Label for Y-axis
plt.show()
```

```python
#changing line style and color
plt.plot(x,y , color = 'green' , linestyle = '--' , marker = 'o')  # Green dashed line with circle markers
plt.show()
```

```python
# 📌 Creating Multiple Lines in the Same Plot
y2 = [1, 4, 6, 8, 10]
plt.plot(x, y, label="Prime Numbers")   # First line
plt.plot(x, y2, label="Random Numbers") # Second line
plt.legend()                            # Show legend to differentiate lines
plt.show()
plt.savefig('plot.png')     # Save current figure as an image file
```

```python
# 📌 Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values)              # Create a bar chart
plt.title("Bar Chart Example")           # Title for the bar chart
plt.show()
```

```python
# 📌 Horizontal Bar Chart
plt.barh(categories, values)             # Create horizontal bar chart
plt.show()
```

```python
sizes = [30, 20, 25, 25]
labels = ['Apples' , 'Banana' , 'Mango' , 'Grapes']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')   # Create pie chart with percentages
plt.title("Fruit Distribution")                   # Title for pie chart
plt.show()

```

```python
# 📌 Scatter Plot
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]
plt.scatter(x, y)                      # Create scatter plot
plt.title("Scatter Plot Example")       # Title for scatter plot
plt.show()
```

```python
# 📌 Histogram
ages = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31, 27]
plt.hist(ages, bins=10)                 # Create histogram with 10 bins
plt.title("Age Distribution")            # Title for histogram
plt.xlabel("Age")                        # X-axis label
plt.ylabel("Frequency")                  # Y-axis label
plt.show()
```

```python
# 📌 Subplots (Multiple Plots in One Figure)
plt.subplot(1, 2, 1)        # 1 row, 2 columns, 1st subplot
plt.plot(x, y)              
plt.title("Plot 1")          # Title for 1st plot
```

```python
plt.subplot(1, 2, 2)        # 1 row, 2 columns, 2nd subplot
plt.bar(categories, values)
plt.title("Plot 2")          # Title for 2nd plot
```

```python
plt.tight_layout()          # Adjust layout to prevent overlap
plt.show()
```

```python
# 📌 Figure Size
plt.figure(figsize=(8, 4))  # Width=8 inches, Height=4 inches
plt.plot(x, y)
plt.show()
```

```python
# 📌 Save a Plot to File
plt.plot(x, y)
plt.savefig('plot.png')     # Save current figure as an image file
plt.show()
```

```python
plt.plot(x, y)
plt.xlim(0, 10)             # Set x-axis limits from 0 to 10
plt.ylim(0, 15)             # Set y-axis limits from 0 to 15
plt.show()
```

```python
# 📌 Grid Lines
plt.plot(x, y)
plt.grid(True)              # Enable grid lines
plt.show()
```

```python
# 📌 Color Maps in Scatter Plot
colors = [10, 20, 30, 40, 50, 60, 70, 80]
plt.scatter(x, y, c=colors, cmap='viridis')   # Color according to value with 'viridis' colormap
plt.colorbar()                                # Show color bar
plt.show()
```

```python
import pandas as pd
```

```python
# 📌 Create a Large Dataset
data = {
    'Year': list(range(2000, 2025)),         # Years from 2000 to 2024 (25 years)
    'Sales': [100 + i*15 + (i%5)*30 for i in range(25)],  # Some pattern in sales data
    'Profit': [20 + i*5 + (i%3)*10 for i in range(25)],   # Some pattern in profit data
    'Expenses': [50 + i*8 + (i%4)*20 for i in range(25)]  # Some pattern in expenses
}
```

```python
# 📌 Create a DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# 🔥 1. Line Plot (Sales over Years)
plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Sales'], marker='o', label='Sales')
plt.title('Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 🔥 2. Bar Plot (Profit per Year)
plt.figure(figsize=(10,5))
plt.bar(df['Year'], df['Profit'], color='green')
plt.title('Profit Over Years')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.show()
```

```python
# 🔥 3. Scatter Plot (Sales vs Profit)
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'], df['Profit'], color='red')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
# 🔥 4. Multiple Lines (Sales, Profit, Expenses)
plt.figure(figsize=(12,6))
plt.plot(df['Year'], df['Sales'], label='Sales')
plt.plot(df['Year'], df['Profit'], label='Profit')
plt.plot(df['Year'], df['Expenses'], label='Expenses')
plt.title('Sales, Profit, and Expenses Over Years')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 🔥 5. Histogram (Distribution of Profit)
plt.figure(figsize=(8,5))
plt.hist(df['Profit'], bins=10, color='purple')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
