# *Linear Plot*

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y)  # Default line
plt.title("Basic Linear Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
#plt.grid(False)
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y, color='red', marker='o', linestyle='--')
plt.title("Styled Linear Plot")
plt.xlabel("X")
plt.ylabel("Y = X^2")
plt.grid(True)
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label="y = x", color="blue")
plt.plot(x, y2, label="y = x^2", color="green")

plt.title("Multiple Linear Lines")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
languages = ['Python', 'Java', 'C++', 'PHP']
popularity = [85, 70, 60, 80]
popularity2 = [20, 30, 40, 50]

plt.bar(languages, popularity, color='red', label='Popularity')
plt.bar(languages, popularity2, bottom=popularity, color='gold', label='Popularity1')

plt.xlabel("Language")
plt.ylabel("Total No.")
plt.title("Stacked Bar Chart")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

languages = ['Python', 'Java', 'C++', 'PHP']
popularity1 = [85, 70, 60, 80]
popularity2 = [20, 30, 40, 50]

x = np.arange(len(languages))
width = 0.4

plt.bar(x - width/2, popularity1, width=width, color='red', label='Popularity')
plt.bar(x + width/2, popularity2, width=width, color='yellowgreen', label='Popularity1')

plt.xticks(x, languages)
plt.xlabel("Language")
plt.ylabel("No.")
plt.title("Side-by-Side Bar Chart")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np


# X-axis labels (languages)
labels = ['Python', 'Java', 'C++', 'PHP']

# Two datasets
data1 = [85, 70, 60, 80]   # Popularity of each language
data2 = [20, 30, 40, 50]   # Another dataset for comparison

# Position of bars
x = np.arange(len(labels))
width = 0.4  # Width of each bar

# Create bars
plt.bar(x - width/2, data1, width=width, color='red', label='Popularity')
plt.bar(x + width/2, data2, width=width, color='yellowgreen', label='Not Popular')

# Add labels and title
plt.xticks(x, labels)
plt.xlabel("Language")
plt.ylabel("No.")
plt.title("Easy Side-by-Side Bar Chart")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
x = [1, 2, 3, 4, 5]               # Positions
y = [50, 80, 30, 60, 40]         # Quantity
colors = ["red", "yellow", "orange", "darkorange", "purple"]
sizes = [200, 300, 150, 250, 180]

plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, edgecolors='black')
plt.xticks(x, fruits)
plt.title("Fruit Quantities in Market", fontsize=14)
plt.xlabel("Fruit Type")
plt.ylabel("Quantity Available")
plt.grid(True)
plt.show()
```

```python
car_names = ["Civic", "Corolla", "BMW", "Audi", "Cultus"]
speed = [120, 115, 180, 200, 100]      # x-axis
mileage = [15, 14, 10, 9, 18]          # y-axis
sizes = [250, 240, 400, 420, 180]      # Engine power visualized
colors = ['blue', 'green', 'black', 'red', 'orange']

plt.scatter(speed, mileage, c=colors, s=sizes, marker='o', edgecolors='black', alpha=0.8)
for i, name in enumerate(car_names):
    plt.text(speed[i]+1, mileage[i]+0.2, name, fontsize=10)

plt.title("Car Comparison: Speed vs Mileage", fontsize=14)
plt.xlabel("Top Speed (km/h)")
plt.ylabel("Mileage (km/l)")
plt.grid(True)
plt.show()
```

```python
students = ["Ali", "hasan", "Sara", "Zain", "Hamza"]
study_hours = [2, 4, 5, 3, 6]         # x-axis
marks = [55, 68, 72, 60, 85]         # y-axis
colors = ['brown', 'blue', 'green', 'purple', 'red']
sizes = [100, 150, 200, 120, 250]

plt.scatter(study_hours, marks, c=colors, s=sizes, marker='*', edgecolors='black') #alpha=0.9
for i, name in enumerate(students):
    plt.text(study_hours[i]+0.1, marks[i]+0.5, name, fontsize=9)

plt.title("Study Hours vs Marks", fontsize=14)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
students = ["Ali", "Hamza", "Sara", "Zain", "Areeba"]
study_hours = [2, 4, 5, 3, 6]
marks = [55, 68, 72, 60, 85]  # used as color values

sizes = [100, 150, 200, 120, 250]

scatter = plt.scatter(study_hours, marks, c=marks, s=sizes, cmap='viridis', edgecolors='black', alpha=0.9)

for i, name in enumerate(students):
    plt.text(study_hours[i]+0.1, marks[i]+0.5, name, fontsize=9)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Marks vs Study Hours with Colorbar")

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label("Marks (Color Encoded)")

plt.grid(True)
plt.show()
```

```python
# Sample data
data = [12, 13, 15, 15, 16, 16, 17, 17, 17, 18, 19, 20, 21, 22, 22, 23, 24, 25]

# Create histogram
plt.hist(data, bins=7, color='yellow', edgecolor='black', bottom = 10, align = "left") # cumulative = -1

#bIn a histogram, bins define the number of groups (intervals or ranges) into which your data is divided.

# Add labels and title
plt.title("Student Ages Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

```python
# Sample data
data = [5, 7, 8, 10, 12, 13, 13, 15, 18, 20, 20, 22, 23, 25, 25, 26]

# Create the histogram with all main parameters
plt.hist(
    data,
    bins=6,                     # Number of bins (intervals)
    color='orange',             # Fill color of bars
    edgecolor='black',          # Border color of bars
    alpha=0.7,                  # Transparency (0 = transparent, 1 = solid)
    histtype='bar',             # Type: 'bar', 'barstacked', 'step', 'stepfilled'
    orientation='vertical',     # 'vertical' or 'horizontal'
    rwidth=0.9,                 # Width of bars relative to bin width
    linewidth=1.2,              # Border line width
    density=False               # If True, shows probability instead of frequency
)

# Add labels and title
plt.title("Complete Histogram Example", fontsize=14)
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.grid(True)

# Show the plot
plt.show()
```

# Stem Plot 

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]

plt.stem(x, y)
plt.title("1. Simple Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x = [0, 1, 2, 3, 4]
y = [3, 8, 1, 10, 5]

(markerline, stemlines, baseline) = plt.stem(x, y)
plt.setp(markerline, color='green', marker='o')
plt.setp(stemlines, color='purple', linewidth=3)
plt.title("2. Colored Stem Plot")
plt.grid(True)
plt.show()
```

```python
x = [10, 15, 7, 12, 5]
y = [1, 2, 3, 4, 5]

# Simulate horizontal by swapping x and y
(markerline, stemlines, baseline) = plt.stem(x, y, orientation='horizontal')
plt.setp(markerline, color='orange')
plt.setp(stemlines, color='gray')
plt.title("3. Horizontal-style Stem Plot")
plt.grid(True)
plt.show()
```

```python
x = range(1, 6)
y = [2, 4, 6, 8, 10]

# Create the stem plot
markerline, stemlines, baseline = plt.stem(
    x, y, linefmt='r--', markerfmt='go', basefmt='k-'
)

# Customize elements
plt.setp(stemlines, linewidth=2)
plt.setp(markerline, markersize=8)

# Add labels and title
plt.title("4. Fully Customized Stem Plot")
plt.xlabel("Index")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

# Box and Whistle 

```python
x = [10, 20, 30, 40, 50, 60]
# 10 min   and 60 maximum 
plt.boxplot(x, widths= 0.4 , labels = ["Python"] )
plt.show()

#notch = False, vert = False
```

```python
data = [[10, 20, 30, 25], [15, 22, 27, 35], [5, 12, 18, 23]]
plt.boxplot(data, labels=["Group A", "Group B", "Group C"])
plt.title("Box Plot with Multiple Groups")
plt.show()
```

```python
data = [
    [5, 7, 8, 12, 14, 15],
    [10, 15, 18, 21, 22, 25],
    [20, 22, 24, 28, 30, 35]
]

plt.boxplot(
    data,
    notch=True,                   # Adds notches for median comparison
    vert=True,
    patch_artist=True,
    labels=['Group 1', 'Group 2', 'Group 3'],
    boxprops=dict(facecolor='lightgreen'),
    flierprops=dict(marker='D', color='purple', markersize=6)
)

plt.title("Multiple Group Box Plots")
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()
```

```python
data = [50, 52, 54, 55, 56, 58, 60, 70, 80, 90, 100]

plt.boxplot(
    data,
    vert=True,                  # vetical box plot
    whis=1.5,                    # Whisker length (1.5 times IQR)
    patch_artist=True,
    boxprops=dict(facecolor='lightcoral'),
    medianprops=dict(color='black'),
    flierprops=dict(marker='x', color='blue', markersize=6)
)

plt.title("Horizontal Box and Whisker Plot")
plt.xlabel("Scores")
plt.grid(True)
plt.show()
```

# Area vs Stack Plot

```python
import numpy as np

x = np.arange(1, 6)
y = np.array([3, 5, 2, 8, 7])

plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='skyblue', alpha=0.5)

plt.title("Area Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 45, 56]
y = [23, 45, 67, 89, 100]  

plt.stackplot(x, y, color='blue', alpha=0.4, labels=['Y values'])
plt.fill_between(x, y, color='skyblue', alpha=0.5)

plt.title("Area vs Stack Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.arange(1, 6)
A = [3, 4, 2, 5, 4]
B = [1, 2, 1, 3, 2]
C = [2, 3, 2, 4, 3]

plt.stackplot(x, A, B, C, labels=['A', 'B', 'C'], colors=['skyblue', 'lightgreen', 'lightcoral'], alpha=0.7)

plt.title("Stack Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# Step PLot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.step(x, y, where='mid', color='purple', linewidth=2)
plt.title("Step Plot - where='mid'")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 15, 10, 20, 25]

plt.step(x, y, where='pre', label="Pre Step", linestyle='--')
plt.step(x, y, where='mid', label="Mid Step", linestyle='-.')
plt.step(x, y, where='post', label="Post Step", linestyle=':')

plt.title("Step Plot Comparison")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

# Subplot

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r')
plt.title("Ascending")

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'g')
plt.title("Descending")

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')
plt.title("Plot 1")

plt.subplot(2, 2, 2)
plt.bar(x, y)
plt.title("Plot 2")

plt.subplot(2, 2, 3)
plt.scatter(x, y, color='green')
plt.title("Plot 3")

plt.subplot(2, 2, 4)
plt.step(x, y, where='mid')
plt.title("Plot 4")

plt.savefig("Subplot")
plt.tight_layout()
plt.show()

```
